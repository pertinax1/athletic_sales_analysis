### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')

```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020_df, sales_2021_df],).reset_index(drop = True)

# Display the combined DataFrame
combined_sales_df.head()
```

```python
# Convert the "invoice_date" to a datetime datatype
# Doing this before the sum() function will cause an error if I try to do sum calculations on the created df
# Created a copy just to perform the action to demosntrate the that I know how to do it
combined_sales_convert_date_df = combined_sales_df.copy()
combined_sales_convert_date_df['invoice_date'] = pd.to_datetime(combined_sales_convert_date_df['invoice_date'])

# Uncomment the line below to see the error
# demonstrate_error = combined_sales_convert_date_df.groupby(['region', 'state', 'city']).sum()['units_sold'].reset_index()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

products_sold_df = combined_sales_df.groupby(['region', 'state', 'city']).sum()['units_sold'].reset_index()
products_sold_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
products_sold_df.head(5)
                            
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_convert_date_df.dtypes
```

#### Using `pivot_table`

```python
# Using pivot_table Show the number products sold for region, state, and city.
products_sold_df = combined_sales_df.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
products_sold_df.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
sales_by_date_df = combined_sales_df.pivot_table(index=['invoice_date'], values='total_sales', aggfunc='sum').reset_index()


# Optional: Rename the "total_sales" column to "Total Sales"
sales_by_date_df.rename(columns={'total_sales': 'Total_Sales'}, inplace=True)


# Show the table.
sales_by_date_df.head(5)
```

```python
sales_by_date_df.dtypes
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
sales_by_date_df['invoice_date'] = pd.to_datetime(sales_by_date_df['invoice_date'])

daily_sales_df = sales_by_date_df.resample('D', on='invoice_date').sum().reset_index()


# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales_df.sort_values(by='Total_Sales', ascending=True, inplace=True)

# Look at the end of the df so I can see something besides zeros.
daily_sales_df.tail(5)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales_df = sales_by_date_df.resample('W', on='invoice_date').sum().reset_index()


# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_sales_df.sort_values(by='Total_Sales', ascending=True, inplace=True)

weekly_sales_df.head(5)
```
